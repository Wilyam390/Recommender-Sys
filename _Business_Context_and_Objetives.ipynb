Nowadays, online platforms are growing massively, but one main problem present in many of them is that they don't properly use user's and product's data. The most impressing fact is that it doesn't happen to only small companies, but large ones as well. 

Take as example our sceneraio. The company currently shows products randomly, without considering different factors such as what customers like, what is popular or reviews. This means that users often see items that they are not really interested in and hence the business loses the chance to improve engagement and satisfaction.

This notebook develops a recommendation system to provide more relevant and personalised suggestions. In order to do so, we explore several recomender approached inclusing non personalised baselines, collaborative filtering, content based recommendation and hybrids. We evaluate them not only how accurate the predictions are but also trough ranking and business perspective to find the most beneficial solution considering the context. 

All these notebooks aims to:
- understand the dataset including its strenghts and limitations
- build several recommendation models with increasing complexity
- compare these different approaches with different evaludation methods
- measure how much better recommendantion are than random product displau
- showcase the results in a way that is suitable for technical and non techincal stakeholders
